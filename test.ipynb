```python
import matplotlib.pyplot as plt
```

```python
epochs = list(range(1, 11))
precisions = [86.38, 89.05, 90.38, 91.27, 91.80, 92.26, 92.61, 92.85, 93.07, 93.25]

plt.figure(figsize=(12, 8))
plt.plot(epochs, precisions)
plt.title('BiLSTM-CRF Training Precisions')
for i in range(10):
    plt.text(epochs[i], precisions[i], f'({precisions[i]}%)')
plt.xlabel('Epoch')
plt.ylabel('Precision(%)')
plt.show()
```
